# Matplotlib 函数介绍

## 1.pyplot.subplots
&emsp;&emsp;plt.subplots()是Matplotlib中一个方便的函数，用于创建一个带有一个或多个子图的图形。它可以同时创建多个子图，并返回一个包含所有子图的Figure对象和一个或多个Axes对象的元组。

&emsp;&emsp;plt.subplots()函数的常见用法是通过指定行数和列数来创建一个包含多个子图的网格。例如，plt.subplots(2, 3)将创建一个包含2行和3列的网格，其中包含6个子图。

&emsp;&emsp;plt.subplots()函数可以接受多个可选参数，用于调整子图和图形的属性。例如，plt.subplots(figsize=(8, 6))将设置图形的大小为8英寸宽和6英寸高。


``pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)``

- ``nrows``：表示子图的行数。
- ``ncols``：表示子图的列数。
- ``sharex``：如果为True，则所有子图共享x轴。默认值为False。
- ``sharey``：如果为True，则所有子图共享y轴。默认值为False。
- ``figsize``：表示图形的大小。默认值为(6.4, 4.8)。
- ``**kwargs``：表示其他关键字参数，用于设置子图的属性，例如标题、坐标轴标签等。
- ``squeeze``：一个布尔值参数，指定是否压缩子图数组的维度，默认为True。
- `` subplot_kw和gridspec_kw``：两个字典参数，分别用于传递给子图对象和GridSpec对象的参数。





&emsp;&emsp;下面是一个简单的示例代码，演示如何使用plt.subplots()创建一个带有一个子图的图形：

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个包含两个子图的图形
fig, axs = plt.subplots(nrows=2, ncols=2)

# 在第一个子图中绘制正弦函数
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
axs[0, 0].plot(x, y)

# 在第二个子图中绘制余弦函数
y = np.cos(x)
axs[0, 1].plot(x, y)

# 在第三个子图中绘制正切函数
y = np.tan(x)
axs[1, 0].plot(x, y)

# 在第四个子图中绘制cot函数
y = 1/np.tan(x)
axs[1, 1].plot(x, y)

# 调整子图间距和整个图形的布局
plt.subplots_adjust(wspace=0.3, hspace=0.3, left=0.1, right=0.9, top=0.9, bottom=0.1)

# 显示图形
plt.show()
```

&emsp;&emsp;在Matplotlib中，Figure对象是表示整个图形的顶层容器。它可以包含一个或多个Axes对象，用于绘制子图。每个Axes对象都包含一个或多个Axis对象，用于管理坐标轴和网格线。

&emsp;&emsp;在使用plt.subplots()函数创建子图时，该函数将返回一个元组(Figure, Axes)，其中Figure对象表示整个图形，Axes对象表示一个或多个子图。因此，fig变量表示返回元组的第一个元素，即整个图形对象。您可以使用fig变量来设置图形的属性，例如大小、标题、背景色等等。

## pyplot.quiver
&emsp;&emsp;在Matplotlib中，ax.quiver()函数可以用来绘制二维矢量场，也就是在二维平面上显示矢量的大小和方向。矢量场在物理、工程、数学等领域都有广泛的应用。

&emsp;&emsp;``pyplot.quiver()``函数接受多个参数，用于指定要绘制的矢量场的位置、大小和方向等信息。其中，最常用的参数包括：

&emsp;&emsp;
- X和Y：表示向量箭头的起始位置的数组或网格数据。
- U和V：表示向量箭头的长度和方向的数组或网格数据。
- C：表示向量箭头颜色的数组或网格数据。
- cmap：指定颜色映射的名称或对象。
- norm：用于规范化颜色映射的对象。
- alpha：向量箭头的透明度。
- pivot：向量箭头的基准点（即起始位置或终止位置）。
- angles：箭头的弧度（即夹角）。
- scale：缩放因子，用于调整箭头的大小。
- width：箭头的宽度。
- headwidth：箭头头部的宽度。
- headlength：箭头头部的长度。
- headaxislength：箭头头部轴的长度。
- minshaft：箭头的最小长度，用于控制箭头是否显示。
- pivot：箭头基准点的位置。

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = plt.axes(projection="3d")


x = torch.arange(-10, 10, 0.1)
y = torch.arange(-10, 10, 0.1)
X ,Y = torch.meshgrid(x,y)
Z = X**2 + Y**2
fx ,fy = torch.gradient(Z)#该函数返回x方向和y方向上的梯度
fig = plt.figure()
ax = plt.axes()
plt.quiver(X,Y,-fx,-fy)

plt.show()
```

```python
# 创建 3D 坐标系
fig = plt.figure()
ax = plt.axes(projection='3d')

num = np.linspace(-10,10,100)
x = np.sin(num)
y = np.cos(num)
z = num

ax.plot3D(x,y,z)#此时绘制的曲线为三维空间上的曲线
```

### pyplot.scatter绘制散点图

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

# 创建 3D 坐标系
fig = plt.figure()
ax = plt.axes(projection='3d')

# 生成随机 3D 散点数据
n = 100  # 数据点数量
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)
# 绘制散点图
ax.scatter(x, y, z)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


# 显示图像
plt.show()
```
