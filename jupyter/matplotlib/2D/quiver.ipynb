# 1.pyplot.quiver
&emsp;&emsp;在Matplotlib中，ax.quiver()函数可以用来绘制二维矢量场，也就是在二维平面上显示矢量的大小和方向。矢量场在物理、工程、数学等领域都有广泛的应用。

&emsp;&emsp;``pyplot.quiver()``函数接受多个参数，用于指定要绘制的矢量场的位置、大小和方向等信息。其中，最常用的参数包括：

&emsp;&emsp;
- X和Y：表示向量箭头的起始位置的数组或网格数据。
- U和V：表示向量箭头的长度和方向的数组或网格数据。
- C：表示向量箭头颜色的数组或网格数据。
- cmap：指定颜色映射的名称或对象。
- norm：用于规范化颜色映射的对象。
- alpha：向量箭头的透明度。
- pivot：向量箭头的基准点（即起始位置或终止位置）。
- angles：箭头的弧度（即夹角）。
- scale：缩放因子，用于调整箭头的大小。
- width：箭头的宽度。
- headwidth：箭头头部的宽度。
- headlength：箭头头部的长度。
- headaxislength：箭头头部轴的长度。
- minshaft：箭头的最小长度，用于控制箭头是否显示。
- pivot：箭头基准点的位置。

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

x = torch.arange(-10, 10, 1)
y = torch.arange(-10, 10, 1)
X ,Y = torch.meshgrid(x,y)
Z = X**2 + Y**2
fx ,fy = torch.gradient(Z)#该函数返回x方向和y方向上的梯度
fig = plt.figure()
ax = plt.axes()
plt.quiver(X,Y,-fx,-fy)

plt.show()
```
