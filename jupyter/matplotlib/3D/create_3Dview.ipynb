```python
import matplotlib.pyplot as plt
import numpy as np
# 创建 3D 坐标系

fig = plt.figure()
ax = plt.axes(projection="3d")


fig = plt.figure()
ax = plt.axes(projection='3d')

num = np.linspace(-10,10,100)
x = np.sin(num)
y = np.cos(num)
z = num

ax.plot3D(x,y,z)#此时绘制的曲线为三维空间上的曲线
```
